```
from pyspark.sql.functions import input_file_name

(spark.read.option("delimiter",";").option("header", True).
 csv("/FileStore/tables/covid_basico.csv").
 withColumn("filename",input_file_name()).
 createOrReplaceTempView("tbl_covid"))
```

```
%sql
SELECT * FROM tbl_covid
```

```
spark.sql("show tables").show(truncate=False)
```

```
%sql
SHOW TABLES
```

```
%sql
SELECT * FROM tbl_covid WHERE REGIAO = 'Norte' OR ESTADO = 'AM'
```

```
%sql
SELECT COUNT(*) as count_norte FROM tbl_covid WHERE REGIAO = 'Norte'
```

```
%sql
DESCRIBE tbl_covid
```

```
%sql
SELECT * FROM tbl_covid WHERE month(data) > 2 and month(data) < 4
```

```
%sql
SELECT * FROM tbl_covid WHERE year(data) > 2020
```

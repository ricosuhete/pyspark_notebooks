```python
from pyspark import SparkConf
from pyspark.sql import SparkSession
import pyspark.sql.functions as F
```

```python
spark = (
    SparkSession.builder
    .master('local[3]')
    .appName('Select e Filter-dataframe')
    .getOrCreate())
```

```python
print(spark)
```

```python
spark
```

```python
df  = (
    spark
    .read
    .format("csv")
    .option("header", "true")
    .option("delimiter", ";")
    .option("infeSchema", "true")
    .load("csv/covid_basico.csv")
)
```

```python
df.printSchema()
```

```python
df.show()
```

```python
df.select("regiao", "estado", "casosNovos").show(10)
```

```python
df.select(F.col("regiao"),F.col("estado"),F.col("casosNovos")).filter(F.col("regiao") == "Norte").show(10)
```

```python
df.select(F.col("regiao"),F.col("estado"),F.col("casosNovos")).filter(df.regiao == "Norte").show(10)
```

```python
df.select(F.col("regiao"),F.col("estado"),F.col("casosNovos")).filter("regiao == 'Norte'").show(10)
```

```python
filtro = F.col("regiao") == "Norte"
df.select(F.col("regiao"), F.col("estado"), F.col("casosNovos")).filter(filtro).show(10)
```

```python
colunas = ["regiao", "estado", "casosNovos"]
df.select(colunas).filter(filtro).show(10)
```

```python
df.printSchema()
```

```python
df.filter("regiao = 'Sul'").filter("estado = 'RS'").show(10)
```

```python
df.filter("regiao == 'Sudeste' AND estado == 'ES'").show(10)
```

```python
df.where("regiao == 'Sudeste' AND estado == 'ES'").show(10)
```

```python
df.filter((F.col("regiao") == "Norte") & (F.col("estado") == "AM")).show(10)
```

```python
df.where((F.col("regiao") == "Norte") & (F.col("estado") == "AM")).show(10)
```

```python
df.where("regiao == 'Sudeste'").where("estado like 'E%'").show(10)
```

```python
df.filter("""regiao in ('Norte', 'Nordeste') and estado like 'A%'""").show(1000)
```

```python
lista_regiao = ["Norte", "Nordeste"]
df.filter(F.col("regiao").isin(lista_regiao)).show(1000)
```
